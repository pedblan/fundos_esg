#### Referências dos fundos ranqueados segundo patrimônio líquido

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from nltk.corpus import stopwords
import nltk

# Baixar as stop words da NLTK
nltk.download('stopwords')

# Carregar as stop words em inglês (ou outra língua conforme necessário)
stop_words = set(stopwords.words('portuguese'))

# Supondo que df é o seu DataFrame e já contém os dados necessários

# Convertendo 'Patrimônio Líquido' para string e limpando para conversão numérica
df['Patrimônio Líquido'] = df['Patrimônio Líquido'].apply(processar_texto).astype(str)
df['Patrimônio Líquido'] = pd.to_numeric(df['Patrimônio Líquido'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Removendo possíveis valores NaN após a conversão
df.dropna(subset=['Patrimônio Líquido'], inplace=True)

# Verificando se o DataFrame não está vazio após remoção
if not df.empty:
    # Dividindo em quartis
    df['Quartil'] = pd.qcut(df['Patrimônio Líquido'], 4, labels=['1º Quartil', '2º Quartil', '3º Quartil', '4º Quartil'])
    
    # Iterando sobre cada quartil
    for quartil in df['Quartil'].unique():
        # Agrupando as referências por quartil
        textos = df.loc[df['Quartil'] == quartil, 'Metodologia ESG - referência'].dropna()
        texto_agrupado = " ".join(texto for texto in textos)
        
        # Filtrando stop words
        texto_agrupado = ' '.join([word for word in texto_agrupado.split() if word.lower() not in stop_words])
        
        # Gerando a word cloud
        wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stop_words).generate(texto_agrupado)
        
        # Mostrando a word cloud
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(f'Word Cloud para {quartil}')
        plt.show()
else:
    print("O DataFrame está vazio após a limpeza dos dados.")

```

#### Clusterização (classificação não supervisionada) conforme resultado da análise de metodologias ESG, chave "Referências"

```python
# Conversão de texto em características numéricas usando TF-IDF
df['Metodologia ESG - referência'] = df['Metodologia ESG - referência'].fillna('')


tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Metodologia ESG - referência'])

# Aplicação de um algoritmo de clusterização
num_clusters = 9 # Defina o número de clusters desejado
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
df['Metodologia ESG - referência - cluster'] = kmeans.fit_predict(tfidf_matrix)

# Ajuste o número de exemplos que deseja ver por cluster
num_exemplos_por_cluster = 10

# Itera sobre cada cluster
for i in range(num_clusters):
    # Filtra o DataFrame para obter apenas as linhas pertencentes ao cluster atual
    cluster_df = df[df['Metodologia ESG - referência - cluster'] == i]
    
    print(f"Exemplos do Cluster {i}:")
    if len(cluster_df) < num_exemplos_por_cluster:
        # Se o cluster tiver menos exemplos do que o solicitado, mostra todos eles
        display(pd.DataFrame(cluster_df['Metodologia ESG - referência'].sample(n=len(cluster_df))))
    else:
        # Caso contrário, seleciona aleatoriamente o número solicitado de exemplos
        display(pd.DataFrame((cluster_df['Metodologia ESG - referência'].sample(n=num_exemplos_por_cluster))))
    print("\n")  # Adiciona uma linha em branco entre os clusters para melhor visualização
```
