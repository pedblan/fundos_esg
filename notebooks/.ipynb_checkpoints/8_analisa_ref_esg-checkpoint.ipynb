# 8. Análise das referências
Faz análises TF-IDF e word cloud na coluna referências do DataFrame, no geral e em quartis específicos. 

```python
import sys
sys.path.append('../src')
import analise_nlp
import pandas as pd
```

```python
# Configurações iniciais do DataFrame
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', lambda x: '%.0f' % x)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
df = pd.read_json('../data/json/dados.json') 
df.set_index('Fundo', inplace=True)
```

## Análise geral das referências

```python
df_ref = df[[
 'Referência ASG', 'Patrimônio Líquido', 'Classe ANBIMA', 'Fundo ESG', 'Gestor',
 'Rentabilidade 12 meses', 'Distribuidor'
 ]].fillna(value='')
df_rel = df[[
 'Relatório ASG', 'Patrimônio Líquido', 'Classe ANBIMA', 'Fundo ESG', 'Gestor',
 'Rentabilidade 12 meses', 'Distribuidor'
 ]].fillna(value='')

```

```python
df_ref['texto_completo'] = df_ref['Referência ASG'].apply(lambda row: ''.join(row))
texto_completo_ref = ' '.join(df_ref['texto_completo'].tolist())
```

```python
texto_unido = " ".join(df_ref['Referência ASG'].astype(str)) 
texto_unido_limpo =  analise_nlp.limpar_texto(texto_unido, analise_nlp.stop_words, return_string=True)
# Prepara word cloud
analise_nlp.gerar_word_cloud(texto_unido_limpo)
```

## Análise das referências por rentabilidade

```python
# Dividir em quartis de patrimônio líquido

df_quartis_pl = analise_nlp.dividir_em_quartis(df_ref, 'Patrimônio Líquido') # 
# Processar textos e gerar word clouds e TF-IDF para cada quartil

for i in range(1, 5):
    quartil = f'{i}º Quartil'
    df_quartil = df_quartis_pl[df_quartis_pl['Quartis Patrimônio Líquido'] == quartil]
    
    texto_quartil = ' '.join(df_quartil['texto_completo'].tolist())
    texto_quartil_limpo_lista = analise_nlp.limpar_texto(texto_quartil, analise_nlp.stop_words)
    texto_quartil_limpo_string = analise_nlp.limpar_texto(texto_quartil, analise_nlp.stop_words, return_string=True)
    
    print(f"Word cloud e value counts para as referências do {quartil}, em termos de patrimônio líquido")
    analise_nlp.gerar_word_cloud(texto_quartil_limpo_string, largura=800, altura=400, cor_fundo='white')
    referencias_quartil = df_quartil['Referência ASG'].value_counts()
    
    print(f"Referências no {quartil}, em termos de patrimônio líquido:")
    display(pd.DataFrame(referencias_quartil))
```

# Grupos de referências

```python
# Agrupar por Gestor e contar as referências
referencias_por_gestor = df_ref.groupby('Gestor')['Referência ASG'].value_counts()

# Imprimir as referências por Gestor
print("Referências por Gestor:")
display(pd.DataFrame(referencias_por_gestor))

```

```python
# Agrupar por Gestor e contar as referências
referencias_por_distribuidor = df_ref.groupby('Distribuidor')['Referência ASG'].value_counts()

# Imprimir as referências por Gestor
print("Referências por Distribuidor:")
display(pd.DataFrame(referencias_por_distribuidor))
```

```python
# Agrupar por Classe ANBIMA e contar as referências
referencias_por_classe_anbima = df_ref.groupby('Classe ANBIMA')['Referência ASG'].value_counts()
print("\nReferências por Classe ANBIMA:")
display(pd.DataFrame(referencias_por_classe_anbima))


```

```python
# Agrupar por Fundo ESG e contar as referências
referencias_por_fundo_esg = df_ref.groupby('Fundo ESG')['Referência ASG'].value_counts()

# Imprimir as referências por Fundo ESG
print("\nReferências por Fundo ESG:")
display(pd.DataFrame(referencias_por_fundo_esg))
```

## Referências por Gestor

```python
# Agrupar por Fundo ESG e contar as referências
referencias_por_gestor = df_ref.groupby('Gestor')['Referência ASG'].value_counts()

# Imprimir as referências por Fundo ESG
print("\nReferências por Gestor:")
display(pd.DataFrame(referencias_por_gestor))
```
