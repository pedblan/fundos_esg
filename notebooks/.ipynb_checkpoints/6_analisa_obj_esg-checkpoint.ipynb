Faz análises TF-IDF e word cloud na coluna objetivos do DataFrame, no geral e em quartis específicos. 

```python
import sys
sys.path.append('../src')
import analise
import pandas as pd
```

```python
# Configurações iniciais do DataFrame
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', lambda x: '%.0f' % x)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
df = pd.read_json('../data/json/dados.json') 
df.set_index('Fundo', inplace=True)
```

## Análise geral dos objetivos

### Filtragem de palavras: supressão de jargão estranho ao contexto ESG e seleção de substantivos e adjetivos

```python
df['Metodologia ESG - objetivo_limpo'] = df['Metodologia ESG - objetivo'].apply(lambda x: analise.limpar_texto(x, stop_words_especiais=True))
```

### TF-IDF

```python
tfidf_resultado = analise.tf_idf(
    dataframe=df,
    coluna_texto='Metodologia ESG - objetivo_limpo',
    coluna_saida='TFIDF',  # Prefixo para colunas de score TF-IDF no DataFrame original
    stop_words_especiais=True  # Se você quer adicionar stop words especiais
)
```

```python
importancia_palavras = tfidf_resultado.sum(axis=0).sort_values(ascending=False) # Calcula escore de palavras
display(pd.DataFrame(importancia_palavras))
```

```python
analise.wc_tfidf(tfidf_resultado)
```
