# 2. Análise exploratória de dados

```python
import sys
sys.path.append('../src')
import pandas as pd
import numpy as np
from IPython.display import display, HTML
import nlp
import grafico
```

```python
df = pd.read_json("../data/json/metadados.json") # Cria um dataframe do Pandas
df.set_index('Fundo', inplace=True) # Estabelece a coluna "Fundo" como índice
df.replace("n/d", np.nan, inplace=True) # Substituindo "n/d" por NaN
```

## Colunas do DataFrame

```python
display(pd.DataFrame(df.columns)) # Colunas do DataFrame
```

## Proporção de fundos classificados como ESG pela Anbima

```python
esg_proportions = df['Fundo ESG'].value_counts(normalize=True) * 100
display(pd.DataFrame(esg_proportions))
```

```python
grafico.elaborar_grafico(esg_proportions, titulo='Proporção de Fundos ESG', x_label='Fundo ESG', y_label='Porcentagem (%)', x_ticks_rotation=0, cores=['green', 'red'])
```

## Proporção de fundos ativos

```python
status = df['Status em 26/02/2024'].value_counts(normalize=True) * 100
display(pd.DataFrame(status))
```

```python
# Em gráfico
grafico.elaborar_grafico(dados=status, titulo='Proporção de Status dos Fundos em 26/02/2024', x_label='Status', y_label='Porcentagem (%)', x_ticks_rotation=45, cores=['green', 'red'])
```

## Proporção de fundos com possibilidade de investimento no exterior

```python
exterior = df['Investimento no exterior'].value_counts(normalize=True) * 100
display(pd.DataFrame(exterior))
```

```python
# Em gráfico
grafico.elaborar_grafico(dados=exterior, titulo='Proporção de Fundos com Investimento no Exterior', x_label='Investimento no Exterior', y_label='Porcentagem (%)', x_ticks_rotation=0, cores=['green', 'red'])
```

## Proporção de fundos segundo tipo estabelecido pela Anbima

```python
tipo_anbima = df['Tipo ANBIMA'].value_counts(normalize=True) * 100
display(pd.DataFrame(tipo_anbima))
```

```python
# Em gráfico
grafico.elaborar_grafico(dados=tipo_anbima, tipo="pie",  titulo='Proporção de Fundos segundo tipologia Anbima', x_label='Tipo Anbima', y_label='Porcentagem (%)', x_ticks_rotation=0, tight_layout=False, cores=['green', 'red', 'yellow', 'purple', 'pink', 'blue'], startangle=0)
```

## Proporção de fundos segundo classe estabelecida pela Anbima

```python
# Em gráfico
grafico.elaborar_grafico(dados=classe_anbima, tipo="pie",  titulo='Proporção de Fundos segundo tipologia Anbima', x_label='Tipo Anbima', y_label='Porcentagem (%)', x_ticks_rotation=0, tight_layout=False, cores=['green', 'red', 'yellow', 'purple', 'pink', 'blue'], startangle=0)
```

## 5 maiores fundos por patrimônio

```python
df_ordenado = df.sort_values(by="Patrimônio Líquido", ascending=False)
top_5 = df_ordenado[["Patrimônio Líquido"]].head(5)
display(HTML(top_5.to_html()))
```

```python
# Filtra o DataFrame para incluir apenas os fundos com status 'Ativo' em 26/02/2024
df_ativos = df[df["Status em 26/02/2024"] == "ATIVO"]
df_ordenado = df_ativos.sort_values(by="Patrimônio Líquido", ascending=False)
top_5 = df_ordenado[["Patrimônio Líquido"]].head(5)
display(HTML(top_5.to_html()))
```

## 5 maiores fundos por rentabilidade

```python
df_ordenado = df.sort_values(by="Rentabilidade 12 meses", ascending=False)
top_5 = df_ordenado[["Rentabilidade 12 meses"]].head(5)
display(HTML(top_5.to_html()))
```

```python
# Filtra o DataFrame para incluir apenas os fundos com status 'Ativo' em 26/02/2024
df_ativos = df[df["Status em 26/02/2024"] == "ATIVO"]
df_ordenado = df_ativos.sort_values(by="Rentabilidade 12 meses", ascending=False)
top_5 = df_ordenado[["Rentabilidade 12 meses"]].head(5)
display(HTML(top_5.to_html()))
```
