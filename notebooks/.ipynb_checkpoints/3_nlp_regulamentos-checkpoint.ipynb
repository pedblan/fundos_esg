```python
import sys
sys.path.append('../src')
import analise_nlp
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from IPython.display import display, HTML
diretorio_txt = "../data/txt"
```

## Carregar todos os regulamentos

```python
documentos = analise_nlp.listar_textos(diretorio_txt) # Lista texto de todos os regulamentos
lista_documentos = documentos.items()
lista_documentos = list(lista_documentos)
```

## Limpar textos

```python
# Procedimento demorado!
textos_limpos = {}
for nome_arquivo, texto_bruto in lista_documentos:
    texto_limpo = analise_nlp.limpar_texto(texto_bruto, analise_nlp.stop_words)  # Use sua lista personalizada de stop words
    textos_limpos[nome_arquivo] = texto_limpo
```

## TF-IDF

```python
lista_de_textos_limpos = textos_limpos.values()
lista_de_textos_limpos = list(lista_de_textos_limpos)
tf_idf = analise_nlp.tfidf(lista_de_textos_limpos)
avg_scores = tf_idf.mean().sort_values(ascending=False)
analise_nlp.pd.set_option('display.max_rows', 500)
display(analise_nlp.pd.DataFrame(avg_scores.head(100)))
```

## Word cloud

```python
textos_para_cloud = " ".join(textos_limpos.values())
analise_nlp.gerar_word_cloud(textos_para_cloud, largura=800, altura=400, cor_fundo='white')
```

## Clusterização

```python
vetores = analise_nlp.obter_embeddings(documentos)
clusters = analise_nlp.clusterizar_textos(vetores, n_clusters=5)
```

```python
analise_nlp.grafico_clusters(vetores, clusters)
```

```python
# Agrupar índices dos documentos por seus clusters
clusters_dict = {}
for indice, cluster in enumerate(clusters):
    if cluster not in clusters_dict:
        clusters_dict[cluster] = []
    clusters_dict[cluster].append(list(vetores.keys())[indice])
```

```python
principais_palavras_por_cluster = analise_nlp.principais_palavras_por_cluster(textos_limpos, clusters_dict, n_palavras=10)
```
